# Day 6

## Part 1: Problem

Problem is specified [here](https://adventofcode.com/2020/day/6).

### Import packages

```python
import csv
```

### Solution

```python
data = []
group = []
with open('06_data.csv', 'r') as file:
    reader = csv.reader(file)
    for row in reader:
        if row == []:
            data.append(group)
            group  = []
        else:
            group = group + list(row[0])
    data.append(group)
```

```python
count = 0
for i in range(len(data)):
    count += len(set(data[i]))
print(count)
```

```python
data = []
group = []
with open('06_data.csv', 'r') as file:
    reader = csv.reader(file)
    for row in reader:
        if row == []:
            data.append(group)
            group  = []
        else:
            group.append(set(list(row[0])))
    data.append(group)
```

```python
data
```

```python
count = 0
for i in range(len(data)):
    if len(data[i]) == 1:
        count += len(data[i][0])
    else:
        count += len(data[i][0].intersection(*data[i][1:]))
print(count)
```
