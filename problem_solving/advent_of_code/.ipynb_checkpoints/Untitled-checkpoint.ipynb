```python
import pandas as pd
```

```python
data = pd.read_csv('05_data.csv', header=None)
data = data.drop(labels=[1, 2, 3, 4, 5, 6], axis=1)
data = data.values.tolist()
print(data)
```

```python
for i in range(len(data)):
    data[i] = list(data[i][0])
```

```python
data
```

```python
binary_code = {'B':1, 'F':0, 'L':0, 'R':1 }
```

```python
biggest = 0
for i in range(len(data)):
    row = 0
    col = 0
    for j in range(0, 7):
        row += binary_code[data[i][j]]*2**(7 - j)
    for j in range(7, 10):
        col += binary_code[data[i][j]]*2**(9 - j)
    res = row * 8 + col
    if res > biggest:
        biggest = res
print(biggest)
```
