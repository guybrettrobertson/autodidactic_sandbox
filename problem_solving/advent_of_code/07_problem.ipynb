```python
data = pd.read_csv('07_data.csv', header=None).values
n = len(data)
data = data.reshape((n))
print(data)
```

```python
bags = []
contains = []
for row in data:
    row = row[:-1]
    bag, contain = row.split(' contain ')
    bag = bag.split(' bag')[0]
    bags.append(bag)
    if contain == 'no other bags':
        contains.append([])
    else:
        contain_split = contain.split(', ')
        contains.append(contain_split)
```

```python
def magic(matrix, n, res, col):
    count = 0
    for i in range(n):
        if matrix.iloc[i, col] == 1:
            if res[i] == -1:
                count += magic(matrix, n, res, i)
                
            else:
                count += res[i]
    res[col] = count
    return None
```
