## Problem 29

Consider all integer combinations of $a^b$ for $2 ≤ a ≤ 5$ and $2 ≤ b ≤ 5$:

$2^2=4$, $2^3=8$, $2^4=16$, $2^5=32$<br>
$3^2=9$, $3^3=27$, $3^4=81$, $3^5=243$<br>
$4^2=16$, $4^3=64$, $4^4=256$, $4^5=1024$<br>
$5^2=25$, $5^3=125$, $5^4=625$, $5^5=3125$

If they are then placed in numerical order, with any repeats removed, we get the following sequence of 15 distinct terms:

$4, 8, 9, 16, 25, 27, 32, 64, 81, 125, 243, 256, 625, 1024, 3125$

How many distinct terms are in the sequence generated by $a^b$ for $2 ≤ a ≤ 100$ and $2 ≤ b ≤ 100$?

#### Solution a

```python
def solve_problem_29a(a_min, a_max, b_min, b_max):
    '''Find the number of distinct terms of a^b in the ranges given.'''
    
    tic = time.perf_counter()
    
    # List containing the terms produced
    terms = []
    
    # Iterate through the possible values of a
    for i in range(a_min, a_max + 1):
        # Iterate through the possible values of b
        for j in range(b_min, b_max + 1):
            a_b = i**j
            # Check if the term is a duplicate
            if a_b not in terms:
                terms.append(a_b)
    
    res = len(terms)
    toc = time.perf_counter()
    
    return res, round(toc - tic, 4)
```

#### Test case

Input:<br>
a_min = 2<br>
a_max = 5<br>
b_min = 2<br>
b_max = 5

Output: 15

```python
a_min = 2
a_max = 5
b_min = 2
b_max = 5

test_case, duration = solve_problem_29a(a_min, a_max, b_min, b_max)
print(test_case)
print('Time taken:', duration, 'seconds')
```
