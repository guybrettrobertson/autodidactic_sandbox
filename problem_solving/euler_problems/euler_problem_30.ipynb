## Problem 30

Surprisingly there are only three numbers that can be written as the sum of fourth powers of their digits:

$1634 = 1^4 + 6^4 + 3^4 + 4^4$<br>
$8208 = 8^4 + 2^4 + 0^4 + 8^4$<br>
$9474 = 9^4 + 4^4 + 7^4 + 4^4$

As $1 = 1^4$ is not a sum it is not included.

The sum of these numbers is $1634 + 8208 + 9474 = 19316$.

Find the sum of all the numbers that can be written as the sum of fifth powers of their digits.

#### Solution a

```python
def solve_problem_30a(power):
    ''''''
    
    tic = time.perf_counter()
    
    powers = {i:i**power for i in range(10)}
    
    special_nums = []
    sum_of_powers_graph = []
    
    upper_limit = 10 ** (power + 1) // 2
    for i in range(2, upper_limit):
        sum_of_powers = sum([powers[int(digit)] for digit in list(str(i))])
        if i == sum_of_powers:
            special_nums.append(i)
        sum_of_powers_graph.append(sum_of_powers)
    
    plt.plot([i for i in range(2, upper_limit)], sum_of_powers_graph)
    plt.plot([i for i in range(2, upper_limit)], [i for i in range(2, upper_limit)])
    plt.xlabel('Number')
    plt.ylabel('Sum of powers of digits')
    plt.show()
    
    res = sum(special_nums)
    
    toc = time.perf_counter()
    
    return res, round(toc - tic, 4)
```

#### Test case

Input: 4

Output: 19316

```python
test_case, duration = solve_problem_30a(4)
print(test_case)
print('Time taken:', duration, 'seconds')
```

#### Final result

Input: 5

Output: 443839

```python
final_result, duration = solve_problem_30a(5)
print(final_result)
print('Time taken:', duration, 'seconds')
```
