# Random walk

## Ants in a grid problem

An ant is trying to get from point A to point B in a grid. The coordinates of point A is (0,0), and the coordinates of point B is (n, n).

Once the ant starts moving, there are four options, it can go left, right, up or down (no diagonal movement allowed). The new point should still be within the boundaries of the n×n grid.

If P is the probability of the ant reaching point BB for a 6×6 grid, use Monte Carlo simulation to compute P. 

## Import packages

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Solution

```python
def take_step(row, col):
    
    rand = np.random.randint(8)
    
    if rand == 0:
        row += 1
        col += 1
    elif rand == 1:
        row += 1
        col -= 1
    elif rand == 2:
        row -= 1
        col += 1
    elif rand == 3:
        row -= 1
        col -= 1
    elif rand == 4:
        row += 1
    elif rand == 5:
        row -= 1
    elif rand == 6:
        col -= 1
    else:
        col += 1
        
    return row, col
```

```python
def ants_in_a_grid(n_ants, grid_size, start_row, start_col):
    '''Simulate and plot random walk of ants in a grid.'''
    
    # Initial conditions
    success = 0 # The counter for the ants which reach B
    fail = 0 # The counter for the ants which do not reach B

    # Iterate through each ant simulation
    for i in range(n_ants):
        row = start_row
        col = start_col
        route = [[start_row, start_col]] # Array to store route taken by each ant

        while True:
            # Update the position of the ant
            row, col = take_step(row, col)
            route = np.vstack([route, [row, col]])

            # Check if the ant is off the grid
            if not (0 <= row <= grid_size and 0 <= col <= grid_size):
                fail += 1
                plt.plot(route[:, 0], route[:, 1], alpha=0.2)
                break

            # Check if the ant has reached the opposite corner
            elif row == grid_size and col == grid_size:
                success += 1
                plt.plot(route[:, 0], route[:, 1], alpha=0.2)
                break

    plt.axis('scaled')
    plt.xlim(0, grid_size)
    plt.ylim(0, grid_size)
    plt.show()
    return success, fail
```

## Examples

```python
ants_in_a_grid(100, 50, 10, 10)
```

```python
ants_in_a_grid(1000, 50, 0, 0)
```

```python
ants_in_a_grid(10, 50, 25, 25)
```
